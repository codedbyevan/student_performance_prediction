```python
# Import lirary
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset from url
url = "https://raw.githubusercontent.com/codedbyevan/student_performance_prediction/refs/heads/main/data/raw/Student_Performance.csv"

df_raw = pd.read_csv(url)
df = df_raw.copy()
```

```python
# View first 5 rows from dataset
df.head(5)
```

```python
df.isnull().sum()
```

```python
# View feature correlation
plt.figure(figsize=(16,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title('Feature Correlation Heatmap')
plt.show()
```
